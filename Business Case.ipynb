```python
import bambi as bmb
import arviz as az
import numpy as np
df['FPD'] = np.where(df['Delay for 1st Installment Payment - Days'] > 3, 1, 0)
df.columns = [c.replace(' ', '_').replace('-', '_') for c in df.columns]
```

```python
# Modelo: Probabilidad de FPD basada en Score y Vertical de Producto
model = bmb.Model("FPD ~ User_Credit_Score + C(Product_Type_Vertical)", data=df, family="bernoulli")
results = model.fit(draws=1000, chains=4)

# Visualización de insights
az.plot_forest(results, var_names=["C(Product_Type_Vertical)"], combined=True)
```
